# Timeseries with Random Forest

This notebook explores using Random Forest to explore a volatile, non-stationary timeseries - the West Australian Wholesale Electricity Market (WEM). Electrical demand is influenced by a range of complex interactions that are evolving over time and Random Forest, even though it doesn't extrapolate, can be useful for the right timseries problems. In this case let's consider seven day ahead forecasitng.

The WEM comprises of the output of a few dozen large generators in the south west of WA. Data from the WEM it does not include embedded generation (e.g. rooftop solar and privately run generators that only push modest amounts of energy back on to the grid), which materialise as reductions in demand, like the "[duck curve](https://en.wikipedia.org/wiki/Duck_curve)". The demand isn't gone, it's simply hidden from the market.

We'll explore a number of sources of explanatory data, but we'll use publicly available data in this notebook. This does restrict our ability to explore the problem - particularly because it means we don't have historic weather forecasts. It's no secret that weather forecasts are imperfect, so any forecast model validated on weather observations will overstate its accuracy ... but we'll just have to pretend we have perfect foresight of the weather to keep the notebook in the public domain.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.ensemble import RandomForestRegressor
```

First up we'll download the last few years of load data from the Australian Energy Market Operator and narrow it down just the Trading Interval (datetime) and Metered Generation data.

I've chosen to drop blank records because they break Random Forest, but there are some cases where you would want to interpolate them.

```python
%%time
df = pd.read_csv('http://data.wa.aemo.com.au/datafiles/load-summary/load-summary-2010.csv', parse_dates=['Trading Date','Trading Interval'])
for i in range(2011, 2020):
  df=df.append(pd.read_csv('http://data.wa.aemo.com.au/datafiles/load-summary/load-summary-'+str(i)+'.csv', 
                           parse_dates=['Trading Date','Trading Interval']))
df=df[['Trading Interval', 'Metered Generation (Total; MWh)']].dropna()
```

A quick check of the data shows that it's in half hour time slots and that generation is measured in megawatt hours per half hour, you'll need to double these values if you're used to working in average megawatts.

```python
df
```

Now let's define a few functions that we'll call as we add new explanatory variables.

**train_forest** takes a data frame, splits it into training and validation sets (based on a given number of records for the validation set), and fits a random forest.

**check_predictions** sets up a 2x2 grid for subplots, and calls predictions_plot for all training and validation results, plus the last week of the training set and the first week of the validation set. The theory is that the Random Forest's accuracy will degrade as the electrical system changes (e.g. additional rooftop solar capacity changes the relationship of sunny days) - but it should be more accurate in the short term.

**predictions_plot** calculates mean absolute error and maximum absolute error, then plots the results

```python
def train_forest(df, split_point):
  train_features = df.drop(['Metered Generation (Total; MWh)'], axis = 1).head(len(df)-split_point)
  train_labels = df['Metered Generation (Total; MWh)'].head(len(df)-split_point)
  val_features = df.drop(['Metered Generation (Total; MWh)'], axis = 1).tail(split_point)
  val_labels = df['Metered Generation (Total; MWh)'].tail(split_point)

  rf = RandomForestRegressor(n_estimators = 50, random_state = 42)
  rf.fit(train_features, train_labels)

  return [rf, train_features, train_labels, val_features, val_labels]

def check_predictions(rf, train_features, train_labels, val_features, val_labels):
  fig,axes  = plt.subplots(nrows=2, ncols=2, figsize=(10,10))
  fig.tight_layout(pad=6.0)
  predictions_plot(rf, train_features, train_labels, axes[0,0])
  predictions_plot(rf, val_features, val_labels, axes[0,1])
  predictions_plot(rf, train_features.tail(7*48), train_labels.tail(7*48), axes[1,0])
  predictions_plot(rf, val_features.head(7*48), val_labels.head(7*48), axes[1,1])

def predictions_plot(rf, features, labels, ax):
  predictions = rf.predict(features)
  errors = abs(predictions - labels)
  title='MAE:' + str(round(np.mean(errors), 2)) + \
        ' MAX_AE:' + str(round(np.max(abs(errors)), 2)) + \
        ' RMSE:' + str(round(np.sqrt(np.mean(np.square(errors))), 2))
  pd.DataFrame({'labels': labels, 
                'predictions': predictions, 
                'Trading Interval': features['Trading Interval']}).plot(title=title, ax=ax, x='Trading Interval')
```

So let's try predicting with just Trading Interval as the only explanatory variable, we'll keep the last year (365 days x 48 intervals per day) as a validation set:

```python
%%time
[rf, train_features, train_labels, val_features, val_labels] = train_forest(df, 365*48)
check_predictions(rf, train_features, train_labels, val_features, val_labels)
```

The results are almost perfect on the training set, but the Random Forest is a flat line in the validation set because the decision trees can't extrapolate a datetime.

So let's try information we can derive from Trading Interval: hour (plus a .5 for the half hour intervals), day of the week, and month of the year. We'll redefine **train_forest** and **predictions_plot** to drop Trading Interval itself as an explanatory variable because it's just going to result in overfitting:

```python
def train_forest(df, split_point):
  train_features = df.drop(['Metered Generation (Total; MWh)'], axis = 1).head(len(df)-split_point)
  train_labels = df['Metered Generation (Total; MWh)'].head(len(df)-split_point)
  val_features = df.drop(['Metered Generation (Total; MWh)'], axis = 1).tail(split_point)
  val_labels = df['Metered Generation (Total; MWh)'].tail(split_point)

  rf = RandomForestRegressor(n_estimators = 50, random_state = 42)
  rf.fit(train_features.drop('Trading Interval', axis=1), train_labels)

  return [rf, train_features, train_labels, val_features, val_labels]

def predictions_plot(rf, features, labels, ax):
  predictions = rf.predict(features.drop('Trading Interval', axis=1))
  errors = abs(predictions - labels)
  title='MAE:' + str(round(np.mean(errors), 2)) + \
        ' MAX_AE:' + str(round(np.max(abs(errors)), 2)) + \
        ' RMSE:' + str(round(np.sqrt(np.mean(np.square(errors))), 2))
  pd.DataFrame({'labels': labels, 
                'predictions': predictions, 
                'Trading Interval': features['Trading Interval']}).plot(title=title, ax=ax, x='Trading Interval')
```

```python
%%time
df.insert(len(df.columns), 'hour', df['Trading Interval'].dt.hour + df['Trading Interval'].dt.minute/60)
df.insert(len(df.columns), 'weekday', df['Trading Interval'].dt.weekday)
df.insert(len(df.columns), 'month', df['Trading Interval'].dt.month)

[rf, train_features, train_labels, val_features, val_labels] = train_forest(df, 365*48)
check_predictions(rf, train_features, train_labels, val_features, val_labels)
```

These results aren't great, but they provide us with an ok baseline model to compare extra explanatory data.

Our model at the moment is a stationary model; it can't change over time, but the electrical system changes every day - new connections, new rooftop solar systems, more efficient applicances, etc. Perhaps adding the year will give us a non-stationary model that doesn't overfit:

```python
%%time
df.insert(len(df.columns), 'year', df['Trading Interval'].dt.year)

[rf, train_features, train_labels, val_features, val_labels] = train_forest(df, 365*48)
check_predictions(rf, train_features, train_labels, val_features, val_labels)
```

The model is beginning to overfit, but that's not so bad - so long as our validation results continue to improve. Problem is, while our mean absolute error on the validation set improved a lot, the max absolute error worsened. Given that we know the electrical system in non-stationary then it makes sense that we should leave the year variable in a continue experimenting.

Next let's try adding weather data. The best (free) data I could find was daily maximum and minimum temperature from the Bureau of Meteorology for the Perth weather station. Obviously we don't have perfect foresight of future weather, but the results should be illustrative.

In the code below we create day of month in order to merge the BOM data, but discard it immediately because having it allows each interval to be unique identified. In other words, it introduces the same overfitting issue as Trading Interval.

```python
%%time
max_temp = pd.read_csv('https://drive.google.com/uc?authuser=0&id=1CHwEwl6A4uqRq_2Q6rtcoICqEMehYR1X&export=download')
min_temp = pd.read_csv('https://drive.google.com/uc?authuser=0&id=1HJV8HtzRlTG_Tn12gUEPemqNt-4E-QWo&export=download')
temp = max_temp.merge(min_temp, left_on=['Year','Month','Day'], right_on=['Year','Month','Day']).dropna()
temp = temp[['Year','Month','Day','Maximum temperature (Degree C)', 'Minimum temperature (Degree C)']]
df.insert(len(df.columns), 'day', df['Trading Interval'].dt.day)
df = df.merge(temp, left_on=['year','month','day'], right_on=['Year','Month','Day']).drop(['day', 'Year', 'Month', 'Day'], axis=1)

[rf, train_features, train_labels, val_features, val_labels] = train_forest(df, 365*48)
check_predictions(rf, train_features, train_labels, val_features, val_labels)
```

The fit on the training data is nearly perfect again, but the validation error improved substantially. Importantly the improvements were on both the mean and max errors.

One persistent theory in electrical forecasting is that heatwaves result much higher consumption than one-off hot days, so next we'll try bringing in the previous day's temperatures.

```python
%%time
temp['max_lag1'] = temp['Maximum temperature (Degree C)'].shift(periods=1)
temp['min_lag1'] = temp['Minimum temperature (Degree C)'].shift(periods=1)
df = df.drop(['Maximum temperature (Degree C)', 'Minimum temperature (Degree C)'], axis=1)
df.insert(len(df.columns), 'day', df['Trading Interval'].dt.day)
df = df.merge(temp, left_on=['year','month','day'], right_on=['Year','Month','Day']).drop(['day', 'Year', 'Month', 'Day'], axis=1)

[rf, train_features, train_labels, val_features, val_labels] = train_forest(df, 365*48)
check_predictions(rf, train_features, train_labels, val_features, val_labels)
```

Even though it's only a small improvement, I find it interesting that the validation results improved by more than the training results.

Let's try adding in the second day before.

```python
%%time
temp['max_lag2'] = temp['Maximum temperature (Degree C)'].shift(periods=2)
temp['min_lag2'] = temp['Minimum temperature (Degree C)'].shift(periods=2)
df = df.drop(['Maximum temperature (Degree C)', 'Minimum temperature (Degree C)', 'max_lag1', 'min_lag1'], axis=1)
df.insert(len(df.columns), 'day', df['Trading Interval'].dt.day)
df = df.merge(temp, left_on=['year','month','day'], right_on=['Year','Month','Day']).drop(['day', 'Year', 'Month', 'Day'], axis=1)

[rf, train_features, train_labels, val_features, val_labels] = train_forest(df, 365*48)
check_predictions(rf, train_features, train_labels, val_features, val_labels)
```

That's a very small improvement. Possibly because heatwaves lasting more than two days are sufficiently uncommon that it's hard for a Random Forest to fit them, possibly because they're sufficiently rare to improve the stats.

One feature of heatwaves that I've noticed is that models often struggle to fit the first day, consumption always seems to be higher than the temperature would suggest. It might be that the forecast of a heatwave causes people to turn on airconditioning that they would leave off on a single hot day.

Let's see if we can improve the model by adding in tomorrow's temperature.

```python
%%time
temp['max_lead2'] = temp['Maximum temperature (Degree C)'].shift(periods=-1)
temp['min_lead2'] = temp['Minimum temperature (Degree C)'].shift(periods=-1)
df = df.drop(['Maximum temperature (Degree C)', 'Minimum temperature (Degree C)', 'max_lag1', 'min_lag1'], axis=1)
df.insert(len(df.columns), 'day', df['Trading Interval'].dt.day)
df = df.merge(temp, left_on=['year','month','day'], right_on=['Year','Month','Day']).drop(['day', 'Year', 'Month', 'Day'], axis=1)

[rf, train_features, train_labels, val_features, val_labels] = train_forest(df, 365*48)
check_predictions(rf, train_features, train_labels, val_features, val_labels)
```

Again the training results barely moved, but the year ahead validation mean error results moved and the week ahead validation max error results came down substantially.

Looking at the validation results it appears that the model has more trouble hitting the low values than the high values. One known driver of low demand is rooftop solar output, so let's see how the model performs if we include solar irradiance (which measures how much light is falling on the earth). The best (free) source that I could find was a daily measurement of solar irradiance for Perth from BOM. Forecasting cloud movement, the primary driver of short term volatility in solar irradiance, is extremely challenging but it's still worth considering whether a daily value would improve the model.

The amount of rooftop solar is constantly growing, approximately 250 MW additional capacity was added in 2019. Hopefully that information will be effectively encoded by the year and month values, and the Random Forest will model the interaction.

```python
solar_output = pd.read_csv('https://drive.google.com/uc?authuser=0&id=1LKKPydF9Owcof9z2hSWjNQjQiy-tCLYa&export=download')
solar_output = solar_output[['Year','Month','Day','Daily global solar exposure (MJ/m*m)']]
df.insert(len(df.columns), 'day', df['Trading Interval'].dt.day)
df = df.merge(solar_output, left_on=['year','month','day'], right_on=['Year','Month','Day']).drop(['day', 'Year', 'Month', 'Day'], axis=1)

[rf, train_features, train_labels, val_features, val_labels] = train_forest(df, 365*48)
check_predictions(rf, train_features, train_labels, val_features, val_labels)
```

That barely moved the dial on the validation set, it would appear that there is something else driving low demand. 

Let's leave new variables for now and have a look at the week of the worst error.

```python
abs_errors = abs(rf.predict(val_features.drop('Trading Interval', axis=1)) - val_labels)
worst_index = abs_errors.tolist().index(max(abs_errors))
fig,axes  = plt.subplots(nrows=1, ncols=1, figsize=(6,6))
predictions_plot(rf, val_features.head(worst_index + 3*48).tail(7*48), val_labels.head(worst_index + 3*48).tail(7*48), axes)
```

This result isn't *terrible*. The worst error appears to be the result of a bit less than expected demand during the middle of the day.

This particular time is towards the end of the validation set, perhaps there has been some change in the behaviour of the system over the last year (like a lot of new rooftop solar systems). Let's try see what would happen if we retrained the model just before this date.

```python
%%time
[rf, train_features, train_labels, val_features, val_labels] = train_forest(df, 365*48 - worst_index + 4*48)
check_predictions(rf, train_features, train_labels, val_features, val_labels)
```

*All* of the validation metrics improved, although alternative metrics (e.g. daily maximum demand accuracy) might tell a different story. Regular retraining of Random Forest models for non-stationary data is important, the more frequently the better. 

Therefore it becomes important to consider how much history is included. Too much will slow training down and not all data will be relevant to today - electricity consumption data from decades ago is unlikely to be helpful in an age of rooftop solar and air conditioners. Too little is also problematic because interesting fringe cases, like heatwaves, occur infrequently and several examples are needed to train a Random Forest.

This notebook is intended to be an incomplete illustration of using Random Forest for electricity forecasting. There are several sources of variation that we haven't considered. Some are public data, like public holidays, and some are a bit harder to get hold of, like planned maintenance of major consumers. There are some data sources that are speculated to signal consumer intent, like the popularity of particular websites and it would be interesting to see whether any of these would significantly improve the result.

It's also important to understand that the model has perfect foresight of weather, which is unrealistic in practise. Before releasing a model like this it's crucial to test using historic weather forecasts to see exactly how well it performs when the weather is uncertain.
